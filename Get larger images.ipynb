```python
import requests            

headers = {'User-Agent': 'My User Agent 1.0',
                       'From': 'bqiongyuyu@gmail.com'}  # This is another valid field

for name, image in zip(gre.df.name.values, gre.df.product_image_url.values):
    response = requests.get(image.replace("grid","hero"), headers=headers)
    if response.status_code == 200:
        with open("flaskexample/static/largerimages/sample{}.jpg".format(name).replace(" ","_"), 'wb') as f:
            f.write(response.content)
    else:
        print("No images found")
```
